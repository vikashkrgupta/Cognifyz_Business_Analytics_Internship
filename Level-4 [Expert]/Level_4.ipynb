**Objective: Analyze common sources participants rely on for investment information.**

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the uploaded csv file
df = pd.read_csv('/content/Data_set.csv')
df.head()
```

```python
# information about dataset
df.info()
```

```python
# extract the source name column from dataset
df1 = pd.read_csv('/content/Data_set.csv', usecols = ['Source'])
df1
```

```python
# description about Source Column
df1.describe()
```

```python
# get information about unique values
df1['Source'].unique()
```

```python
# Analyse Common Information source using bar chart

plt.bar(df1['Source'].value_counts().index, df1['Source'].value_counts().values, edgecolor = 'k', color = 'red', alpha = 0.7)
plt.title('Analyse Common Information Sources', color ='blue', fontsize = 20)
plt.xticks(rotation = 90)
plt.xlabel('Source Participants', fontsize = 15)
plt.ylabel('Counts', fontsize=15)
plt.show()
```

```python
# Analyze the data to identify and summarize the most common sources participants rely on

for Source, count in df1['Source'].value_counts().items():
  print(f"-{Source}: {count}")

most_common_source = df1['Source'].value_counts().head(1)
print("Most Common Source is ", most_common_source.index[0])
```
