## Most Preferred Investment Avenue
**Objective: Identify the most preferred investment avenue.**

```python
# read the investments columns
df2 = pd.read_csv('/content/Data_set.csv', usecols = ['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold'])
df2
```

```python
# Identify the average investment for each avenue
df2.mean(axis=0)
```

```python
# create bar chart  of Average Investment for each Avenue
df2.mean().plot(kind='bar', edgecolor = 'k', color ='orange', alpha = 0.7)
plt.title('Bar Chart of Average Investment for each avenue', color ='red', fontsize = 15)
plt.xlabel('Investment Avenues', color = 'green', fontsize = 12)
plt.ylabel('frequency', color = 'green', fontsize = 12)
plt.show()
```

```python
# Determine the investment avenue with the highest frequency or occurrence.
highest_mean_value = df2.mean().max()
highest_mean_column_name = df2.mean().idxmax()

print(f"Highest Frequency:- {highest_mean_value}")
print(f"Highest Investment Avenue:- {highest_mean_column_name}")
```
