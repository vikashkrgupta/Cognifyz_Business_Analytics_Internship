## Gender Distribution
**Visualize gender distribution in the dataset**

```python
# Extract Gender Information
df1 = pd.read_csv('/content/Data_set.csv', usecols = ['gender'])
df1
```

```python
# Visualize the gender distribution using a bar chart
df1['gender'].value_counts().plot(kind='bar', edgecolor = 'k', color = 'orange')
plt.title('Gender Distribution', color ='red', fontsize =20)
plt.xlabel('Gender', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.show()
```
