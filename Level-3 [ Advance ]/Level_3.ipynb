## Reasons for Investment
**Objective: Analyze and summarize reasons for investment choices.**

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read dataset
df = pd.read_csv('/content/Data_set.csv')
df.head()
```

```python
# get information about the table
df.info()
```

```python
# extract the columns of reason
df1 = pd.read_csv('/content/Data_set.csv', usecols = ['What are your savings objectives?'])
df1
```

```python
# get description of reason column
df1.describe()
```

```python
# Extract all unique values in Reason column
unique_values = df1['What are your savings objectives?'].unique()
print(unique_values)
```

```python
# Create bar graph for unique value with their counts in columns of reasons

value_counts = df1['What are your savings objectives?'].value_counts()

# bar chart
plt.bar(value_counts.index, value_counts.values, edgecolor ='k', color ='brown')
plt.xlabel('Reasons for Investment', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.title('Summary of Reasons for Investment', color = 'red', fontsize = 20)
plt.show()
```

```python
#Summary for the reason for invesment from column of reason

top_reasons = df1['What are your savings objectives?'].value_counts().head(3)

# Print the summary
print("Top 3 reasons for investment:")
for reason, count in top_reasons.items():
  print(f"- {reason}: {count}")
```

# **Task-6**
## Savings Objectives
**Objective: Identify and present main savings objectives.**

```python
# extract the column of objective
df2 = pd.read_csv('/content/Data_set.csv', usecols=['Objective'])
df2
```

```python
# decribe of objective columns
df2.describe()
```

```python
# Extract all unique values from objective column
unique_value = df2['Objective'].unique()
print(unique_value)
```

```python
# Create bar graph for unique value with their counts in column of Objective

value_count = df2['Objective'].value_counts()

# bar chart
plt.bar(value_count.index, value_count.values, edgecolor ='k', color ='Green', alpha = 0.5)
plt.xlabel('Saving Objective', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.title('Summary of Saving Objective ', color = 'red', fontsize = 20)
plt.show()
```

```python
# Summary for the saving objectives from the column of objective

top_objectives = df2['Objective'].value_counts().head(3)

# Print the summary
print("Top 3 saving objectives:")
for objective, count in top_objectives.items():
  print(f"- {objective}: {count}")
```
